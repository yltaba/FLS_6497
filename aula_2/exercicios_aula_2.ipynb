* Crie uma função (ou um código) para sortear 1000 observações do banco de dados climáticos, calcular a média de umidade (humidity).

```python
def get_humidity_sample_avg(df, column='humidity'):
    amostra = df[column].sample(1000)
    return amostra.mean()
```

# c) Modelos lineares
* Modifique a função criada anteriormente para, depois de sortear 1000 observações do banco, rodar um modelo de regressão linear para predizer valores de humidity e extrair o r2 do modelo.\
Dica:\

```python
def get_humidity_reg_score(df):
    y_amostra = df['humidity'].sample(1000)
    X = df[['maximum_temprature', 'cloud_coverage', 'minimum_temprature']].sample(1000)
    reg = LinearRegression().fit(X, y_amostra)
    return reg.score(X, y_amostra), reg.predict(X)
```

```python
resultados_humidity = get_humidity_reg_score(c)
r2_reg_humidity = resultados_humidity[0]
humidity_pred = resultados_humidity[1]
c_amostra = c.sample(1000)
c_amostra['humidity_pred'] = humidity_pred
```

```python
(
    sns.lmplot(
        x='humidity', 
        y='humidity_pred', 
        data=c_amostra, 
        fit_reg=True, 
        line_kws={'color': 'red'}, 
        scatter_kws={'alpha':0.3}
        )
        .set(title='Valores previstos de humidity vs Valores reais')
        )
print(f'O R² deste modelo é igual a {r2_reg_humidity.round(3)}')
```

### Modelo linear com cloud_coverage, minimum_temperature e maximum_temperature como preditores

```python
y = c.humidity
X = c[['cloud_coverage', 'minimum_temprature', 'minimum_temprature']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lin_reg = LinearRegression().fit(X_train, y_train)
r2_lin_reg = lin_reg.score(X_test, y_test)
print(f'O modelo linear tem R² igual a {r2_lin_reg.round(3)}')
```

### Modelo linear polinomial quadrático com cloud_coverage, minimum_temperature e maximum_temperature como preditores

```python
poly_reg = make_pipeline(PolynomialFeatures(degree=2),LinearRegression())
poly_reg = poly_reg.fit(X_train, y_train)
r2_poly_reg = poly_reg.score(X_test, y_test)
print(f'O modelo polinomial tem R² igual a {r2_poly_reg.round(3)}')
```

### Comparando valores previstos com valores observados:

```python
c['y_humidity_lin_pred'] = lin_reg.predict(X)
c['y_humidity_poly_pred'] = poly_reg.predict(X)
```

```python
(
    sns.lmplot(
        x='humidity', 
        y='y_humidity_lin_pred', 
        data=c, 
        fit_reg=True, 
        line_kws={'color': 'red'}, 
        scatter_kws={'alpha':0.3}
        )
        .set(title='Valores previstos de Humidity pela regressão linear vs Valores reais de Humidity')
        )
```

```python
(
    sns.lmplot(
        x='humidity', 
        y='y_humidity_poly_pred', 
        data=c, 
        fit_reg=True, 
        line_kws={'color': 'red'}, 
        scatter_kws={'alpha':0.3}
        )
        .set(title='Valores previstos de Humidity pela regressão polinomial vs Valores reais de Humidity')
        )
```
